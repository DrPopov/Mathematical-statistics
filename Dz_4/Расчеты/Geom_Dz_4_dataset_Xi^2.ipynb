```python
import timeit
import matplotlib.pyplot as plt
import random
import numpy as np
import pylab
import math 
import scipy.stats as st
import pandas as pd
```

```python
fin = open("full.txt")
fd = fin.read()
```

```python
seq = fd.replace('\n', '') #наш ДНК
```

```python
n = len(seq) # длина 
n
```

```python
# вдвигаемся по списку, пока не найдем нужным нам нуклеотид(Т)
seq_list = [] # список, который будет содеражть кол-во шагов до встречи нужного нуклеотида
i = 1 # счетчик

for elem in seq:

    if elem != 'T':
        i += 1
    else:
        seq_list.append(i)
        i = 1

my_set = list(set(seq_list))
amount = [] 

for i in my_set:
    a = seq_list.count(i)
    amount.append(a)
```

```python
def table(): 
    x_mean = np.mean(my_set) # выборочное среднее 
    
    print(x_mean)
    p = 1 / x_mean # оценка парметра p
    
    print('Оценка параметра равна: ',np.around(p, 3) )
    
    list_p_i = [] # теоретические вероятности 

    for elem in my_set:
        p_i = p * (1 - p)**elem
        list_p_i.append(np.around(p_i, 8))

    list_n = [] # теоретические частоты

    for elem in list_p_i:
        list_n.append(n * elem)



    print(list(map(len, [seq_list, amount, list_p_i, list_n])))

    data = {'k': my_set,
           'n_k': amount,
           'p_k': list_p_i,
           'n_k_2': list_n}


    #print(m)
    #print(len(m))
    df = pd.DataFrame(data)
    
    df['Мера'] = np.around((df['n_k'] - df['n_k_2'])**2 / (df['n_k_2']), 3)
    df.loc['Сумма']= df.sum()
    display(df)
    
    print('Кртичиеская точка: ', df['Мера']['Сумма'])
```

```python
table()
```
