# Экспоненциальное распределение Хи2 для сложной гипотезы

```python
import random
import math
from math import e
import numpy as np
import pandas as pd
```

```python
def make_expon_viborka(n, lam = 0.8):
    lst = []
    for i in range(n):
        r = random.random()
        x = -math.log(r,math.e) / lam
        lst.append(float('{:.3f}'.format(x)))
    
    return lst
```

```python
#n = int(input()) # объем выборки
n = 100
```

```python
my_list = make_expon_viborka(n)
```

```python
my_set = list(set(my_list))
amount = []

for i in my_set:
    a = my_list.count(i)
    amount.append(a)


new_list = sorted(my_list)

number = 5 # количество чисел в интервале
count = 1 # cчетчик сколько чисел в интервале

final_interval = []
final_interval.append(new_list[0])

for i in new_list:
    if count == number:
        final_interval.append(i + 0.01)
        count = 0
    count += 1
#print(count)

print(final_interval)

final_interval_2 = final_interval[1:]
final_interval.pop(-1)
#final_interval_2.append(new_list[-1])

#print(new_list)
print(final_interval)
print(final_interval_2)



n_list = [number] * len(final_interval)
if len(n_list) != len(final_interval):
    n_list.append(count)


#print(list(map(len, [final_interval, final_interval_2, n_list])))
#print(new_list)
```

# Сдеалаем свобдную таблицу

```python
def table(): 
    x_mean = np.mean(my_set) # выборочное среднее 
    
    p = 1 / x_mean # оценка парметра p
    
    print('Оценка параметра равна: ',np.around(p, 3) )
    

    data = {'x_k': final_interval,
           'x_k+1': final_interval_2,
            'n_k': n_list}


    df = pd.DataFrame(data)
    
    df['P_k'] = np.around(e**(-p * df['x_k']) - e**(-p * df['x_k+1']), 3)
    df['n_k_2'] = np.around(n * df['P_k'], 3)
    df['Мера'] = np.around((df['n_k'] - df['n_k_2'])**2 / (df['n_k_2']), 3)
    df.loc['Сумма']= df.sum()
    display(df)
```
