# Geometric distribution

```python
import timeit
import matplotlib.pyplot as plt
import random
import numpy as np
import pylab
import math 


#GeomB -  Первая модель моделирования
def geoMB(n,p):
    lst = []
    for t in range(n):
        i = 0
        k = 0
        while i == 0:
            a = random.random()
            if a < p:
                i = 1
            else:
                k += 1
        e = k
        lst.append(e)
        
    lst.sort()
    
    pylab.plot(lst,  label = 'Через испытания Бернулли')
    pylab.legend(title = "n = %s, p = 0.1" %n)


    
# GeomIS - вторая модель моделирования
def geomIS(n, p):
    lst = []
    for i in range(n):
        r = p
        s = r
        k = 0
        q = 1 - p
        a = random.random()
        while a > s:
            k += 1
            r *= q
            s += r
        e = k
        lst.append(e)

    lst.sort()
    
    pylab.plot(lst,  label = 'Методом обратных функций')
    pylab.legend(title = "n = %s, p = 0.1" %n)


    
# GeomDIM - третья модель моделирования
def geomDIM(n, p):
    lst = []
    for i in range(n):
        a = random.random()
        e = math.floor(math.log(a,math.e) / math.log(1 - p,math.e) )
        lst.append(e)

    lst.sort()
    
    pylab.plot(lst,  label = 'Прямой метод обратных функций')
    pylab.legend(title = "n = %s, p = 0.1" %n)


# numpy random
def geo_pyt(n,p):
    lst = []
    for i in range(n):
        a = np.random.geometric(p, size=None)
        lst.append(a)

    lst.sort()
    
    pylab.plot(lst, label =  'NumPy random')
    pylab.legend(title = "n = %s, p = 0.1" %n)
```

# Exponential distibution

```python
# my own func
def expon(lam, n):
    lst = []
    for i in range(n):
        r = random.random()
        x = -math.log(r,math.e) / lam
        lst.append(x)
    lst.sort() 
    
    pylab.plot(lst, label = "my own func")
    pylab.legend(title = 'lambda = %s, n = %d' %(lam, n))


#Встроенная функция экспоненциального распределения
def py_expon(lam, n):
    lst =[]
    for i in range(n):
        a = random.expovariate(lam)
        lst.append(a)

    lst.sort()
    pylab.plot(lst, label ="numpy random")
    pylab.legend(title = 'lambda = %s, n = %d'%(lam,n))
```

Графики

```python
#Строим графики для 1000 точек:
expon(0.5, 1000)
py_expon(0.5, 1000)
```

```python
#Строим графики для 5000 точек:
expon(0.5, 5000)
py_expon(0.5, 5000)
```

```python
#Строим графики для 1000000 точек:
expon(0.5, 1000000)
py_expon(0.5, 1000000)
```
